```python
import csv
import tweepy
import ssl
import re
import time

#tweet collection seconds
collection_time = 3600 
```

```python
# Oauth keys (for account protection, the below keys are not real)
consumer_key = "--"
consumer_secret = "--"
access_token = "--"
access_token_secret = "--"

# Authentication with Twitter
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

target = open("Data.csv", "a+", newline='', encoding="UTF-8")
w = csv.DictWriter(target, fieldnames=['time', 'user', 'text'])
w.writeheader()
```

```python
class StreamListener(tweepy.StreamListener):
    def on_status(self, status):
        try:
            if status.retweeted_status:
                return
        except:
            name = status.user.screen_name
            text = status.text
            created = status.created_at
            print(text)
            w.writerow({'time': created, 'user': name, 'text': text})
    def on_error(self, status_code):
        if status_code == 420:
            return False
```

```python
stream_listener = StreamListener()
stream = tweepy.Stream(auth=api.auth, listener=stream_listener)
stream.filter(track=["realdonaldtrump", "joebiden"], is_async=True)
time.sleep(collection_time)
stream.disconnect()
target.close()
```
