```python
import preprocessor as p
import pandas as pd
import re
from googletrans import Translator
from gensim.parsing.preprocessing import remove_stopwords
translator = Translator()
```

```python
trump = pd.read_csv("Trump_.csv")
```

```python
p.set_options(p.OPT.URL, p.OPT.EMOJI, p.OPT.RESERVED, p.OPT.MENTION)
```

```python
trump = trump.drop(columns=['time', 'user'])
```

```python
trump = trump.astype(str)
trump.dropna(subset=["text"],axis=0,inplace=True)
trump.drop_duplicates(subset ="text", 
                   keep = 'first', inplace = True) 
```

```python
def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    text = re.sub(r'#', '', text)
    text = translator.translate(text).text
    text = remove_stopwords(text)
    print('ok')
    return text
```

```python
trump['text'] = trump.apply(preprocess_tweet, axis=1)
```
